```python
images_0 = '/home/user/data/data/ML/train/target_0'
images_1 = '/home/user/data/data/ML/train/target_1'

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```
